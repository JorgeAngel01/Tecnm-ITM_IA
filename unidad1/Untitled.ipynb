```python
import os
from keras.preprocessing.image import ImageDataGenerator, img_to_array, load_img

# Set the path to your original images and the directory to save augmented images
original_data_path = "../../flowers5/"
save_path = "../../dataAugmented"

# Create the ImageDataGenerator for augmentation
datagen = ImageDataGenerator(
    rotation_range=30,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

# Loop through each class directory in the original data path
for class_name in os.listdir(original_data_path):
    class_path = os.path.join(original_data_path, class_name)
    save_class_path = os.path.join(save_path, class_name)

    # Create a new directory for the class in the save path
    os.makedirs(save_class_path, exist_ok=True)

    # Loop through each image in the class directory
    for filename in os.listdir(class_path):
        img_path = os.path.join(class_path, filename)

        # Load the image
        img = load_img(img_path)
        x = img_to_array(img)
        x = x.reshape((1,) + x.shape)  # Reshape to (1, height, width, channels)

        # Generate augmented images and save them
        i = 0
        for batch in datagen.flow(x, batch_size=1, save_to_dir=save_class_path, save_prefix='aug', save_format='jpeg'):
            i += 1
            if i >= 6:  # Generate 5 augmented images per original image
                break

print("Data augmentation complete.")
```

```python
import os
from PIL import Image  # Import the Image class from the PIL module
from keras.preprocessing.image import ImageDataGenerator, img_to_array, load_img
from skimage import exposure

# Set the path to your original images and the directory to save augmented images
original_data_path = "../../a/Nueva carpeta/"
save_path = "../../dataAugmented"

# Create the ImageDataGenerator for augmentation
datagen = ImageDataGenerator(
    rotation_range=30,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='reflect'
)

# Loop through each class directory in the original data path
for class_name in os.listdir(original_data_path):
    class_path = os.path.join(original_data_path, class_name)
    save_class_path = os.path.join(save_path, class_name)

    # Create a new directory for the class in the save path
    os.makedirs(save_class_path, exist_ok=True)

    # Loop through each image in the class directory
    for filename in os.listdir(class_path):
        img_path = os.path.join(class_path, filename)

        # Load the image
        img = load_img(img_path)
        x = img_to_array(img)
        x = x.reshape((1,) + x.shape)  # Reshape to (1, height, width, channels)

        # Generate augmented images and save them
        i = 0
        for batch in datagen.flow(x, batch_size=1, save_to_dir=save_class_path, save_prefix='aug', save_format='jpeg'):
            # Additional processing for brightness and contrast
            augmented_image = exposure.adjust_gamma(batch[0], gamma=1.5)  # Adjust brightness
            augmented_image = exposure.rescale_intensity(augmented_image, in_range='image', out_range=(0, 255))  # Adjust contrast

            # Save the augmented image
            img_filename = os.path.join(save_class_path, f'aug_{i}_{filename}')
            img = Image.fromarray(augmented_image.astype('uint8'))
            img.save(img_filename)

            i += 1
            if i >= 10:  # Generate n augmented images per original image
                break

print("Data augmentation complete.")
```

```python
import os
import re
import matplotlib.pyplot as plt
from skimage import io, transform, img_as_ubyte

dirname = os.path.join(os.getcwd(), '../../a/Nueva carpeta/paradisebirds/')
imgpath = dirname + os.sep 
output_path = "../../a/dataset/"  # Replace with the desired output directory

small_size = (256, 256, 3)  # Adjust the size as needed

for root, dirnames, filenames in os.walk(imgpath):
    for filename in filenames:
        if re.search("\.(jpg|jpeg|png|bmp|tiff)$", filename):
            filepath = os.path.join(root, filename)
            image = plt.imread(filepath)
            
            if len(image.shape) == 3:
                # Resize the image to the target size
                resized_image = transform.resize(image, small_size)
                # Convert back to uint8 for saving
                resized_image = img_as_ubyte(resized_image)
                
                # Save the resized image to the output directory
                output_filepath = os.path.join(output_path, filename)
                io.imsave(output_filepath, resized_image)

print("Resized images saved to:", output_path)
```
